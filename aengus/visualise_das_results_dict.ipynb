# DAS losses for persona

```python
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
import torch


results = torch.load("/root/quirky_interp/aengus_das_results/DAS_inv-seq-True_inv-pers-False_subdim-1_Mar13-1334-38/results.pt")

# Extracting test_persona_loss values
layers = sorted(results.keys())
layers.remove(17)
positions = range(-7, 0)  # From -7 to -1
data = np.array([[results[layer][pos]['test_persona_loss'] for pos in positions] for layer in layers])

# Plotting
fig, ax = plt.subplots()
cax = ax.matshow(data, interpolation='nearest', cmap='viridis', norm=colors.LogNorm())
fig.colorbar(cax)

ax.set_xticklabels([''] + list(positions))
ax.set_yticklabels([''] + list(layers))
ax.set_xlabel('Position')
ax.set_ylabel('Layer')
ax.set_title('Test Persona Loss Heatmap')

plt.show()
```
